````python
from sklearn.model_selection import  train_test_split
````

fitting the model and evaluating

```
# This is formatted as code
```



````python
maxdepths=[2,3,4,5,6,7,8,9,10,15,20,25,30,35,40,45,50,55]
````

````python
trainaccuracy=np.zeros(len(maxdepths))
````

````python
trainaccuracy
````

````python
testaccuracy=np.zeros(len(maxdepths))
````

````python
testaccuracy
````

````python
index=0
````

````python
for depth in maxdepths:
  clf=tree.DecisionTreeClassifier(max_depth=depth)
  clf=clf.fit(X_train,y_train)
  y_predictTrain=clf.predict(X_train)
  y_predictTest=clf.predict(X_test)
  trainaccuracy[index]=accuracy_score(y_train,y_predictTrain)
  testaccuracy[index]=accuracy_score(y_test,y_predictTest)
  index+=1
````

````python
trainaccuracy
````

````python
testaccuracy
````

````python
import matplotlib.pyplot as plt
import matplotlib_inline
````

````python
plt.plot(maxdepths,trainaccuracy, 'ro-', maxdepths,testaccuracy, 'bv--')
plt.legend(['trainaccuracy','testaccuracy'])
plt.xlabel('maxdepths')
plt.ylabel('accuracy')
````
